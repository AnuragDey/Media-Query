# Media Query

```python
# 1. Create a simple webpage with a header, a main content section, and a footer. Apply a media query that changes the background color of the header to blue when the screen width is less than 600px.
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
    <link rel="stylesheet" href="style (1).css">
</head>
<body>

    <header>
        <h1>Reponsive Header</h1>
    </header>

    <section>
        Main Content
    </section>

    <footer>
        Footer
    </footer>

</body>
</html>
```

```python
header {
    text-align: center;
    background-color: lightgray;
    padding: 5px;
}

section {
    padding: 5px;
}

footer {
    padding: 5px;
}

@media screen and (max-width: 600px) {
    header {
        background-color: blue;
    }
}
```

```python
# 2. Create an image gallery with three images in a row. Use media queries to adjust the layout to two images in a row for screens smaller than 800px and one image in a row for screens smaller than 500px.
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
    <link rel="stylesheet" href="style (2).css">
</head>
<body>
    <div class="imagegallery">
        <img src="https://fastly.picsum.photos/id/313/200/300.jpg?hmac=7_pHHv7TyDti50LaIVPvCPkdWDgM0tcX3ViCIWruEDQ">
        <img src="https://fastly.picsum.photos/id/982/200/300.jpg?hmac=xTKU9lOLW4f1_EmdynREJoYXZ7jXQUnwUNcYa6Nf4WA">
        <img src="https://fastly.picsum.photos/id/664/200/300.jpg?hmac=Ov1G0ZpIuC3e0t33HURn4DPJFK6o7bz602P6M-o_SDc">
    </div>
</body>
</html>
```

```python
.imagegallery {
    display: flex;
    justify-content: space-around;
    padding: 10px;
}

img {
    margin: 10px;
    height: 480px;
    width: 360px;
}

@media screen and (max-width: 800px) {
    .imagegallery{
        display: grid;
        grid-template-columns: repeat(2,1fr);
    }
}

@media screen and (max-width: 500px) {
    .imagegallery{
        display: block;
    }
}
```

```python
# 3. Create a navigation bar with five links. Apply a media query that converts the navigation into a vertical list for screens smaller than 768px and also changes the background color to gray.
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>
    <link rel="stylesheet" href="style (3).css">
</head>
<body>
    <header>
        <nav>
            <ul>
                <li><a href="Task 3.html">Home</a></li>
                <li><a href="">About</a></li>
                <li><a href="">Services</a></li>
                <li><a href="">Portfolio</a></li>
                <li><a href="">Contact</a></li>
            </ul>
        </nav>
    </header>
</body>
</html>
```

```python
* {
    margin: 0;
    padding: 0;
}

header {
    height: 50px;
    display: flex;
    justify-content: center;
    align-items: center;
    margin: 5px;
    background-color: black;
}

ul {
    list-style-type: none;
    display: flex;
}

a {
    text-decoration: none;
    color: white;
    margin: 0px 100px;
    font-weight: bolder;
}

@media screen and (max-width: 768px) {
    header {
        height: 200px;
        background-color: gray;
    }

    ul {
        list-style: none;
        display: flex;
        flex-direction: column;
        justify-content: space-around;
    }

    a {
        display: block;
        padding: 10px 15px;
        text-decoration: none;
        color: white;
    }
}
```
